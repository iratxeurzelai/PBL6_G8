# Mutrition Similar Food Finder

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv('Recetas_procesadas.csv')
```

```python
similarityMatrix=cosine_similarity(df[['calories', 'protein', 'fat', 'sodium']])
```

```python
similarityMatrix
```

```python
similar_food = list(enumerate(similarityMatrix[3]))
```

```python
sorted_similar_food = sorted(similar_food, key=lambda x:x[1], reverse=True)
sorted_similar_food
```
